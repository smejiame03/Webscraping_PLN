Proceso de **Webscraping** y **Procesamiento de Lenguaje Natural(PLN)** para extraer conocimientos generales en común que están siendo requeridos en el mercado laboral de 3 empresas de la ciudad de Medellín, para el cargo de Ingeniero de Sistemas en el área de desarrollo.

```python
#Import needed libraries
from bs4 import BeautifulSoup as bs #webscraping
import requests as rq 
import nltk #Procesamieto de lenguaje natural
import nltk.corpus #stopwords
from nltk.probability import FreqDist #palabras que más se repiten
```

1. **WebScraping** para la primera empresa (Pragma)

```python
url='https://www.pragma.co/carrera/vacantes/desarrollador-frontend'
page=rq.get(url).text
soup=bs(page) #se realiza para mejorar la forma de la lectura de los datos
soup #archivo de tipo HTML 
```

3. Transformaciones necesarias para obtener el texto.

```python
caracteristicas = []
lis = soup.find_all('li', {"aria-level": 1})[-11:]

for li in lis:
    caracteristicas.append(li.span.text)
caracteristicas
```

WebScraping para la segunda empresa (IAS)

```python
url1='https://www.ias.com.co/soy_programador'
page=rq.get(url1).text
soup1=bs(page) 
soup1 
```

```python
caracteristicas1 = []
lis = soup1.find_all('li', class_='list-group-item')[-3:]
for li_element in lis:
    text = li_element.text.strip()
    caracteristicas1.append(text)
caracteristicas1
```

WebScraping para la tercera empresa (IBM)

```python
url2='https://careers.ibm.com/job/18106242/desarrollador-fullstack-remote/?codes=WEB_SEARCH_LA'
page=rq.get(url2).text
soup2=bs(page) #se realiza para mejorar la forma de la lectura de los datos
soup2 #archivo de tipo HTML 
```

```python
caracteristicas2 = []
text = soup2.select('.jd-description p')[-2].get_text(strip=True)
caracteristicas2.append(text)
caracteristicas2
```

Concatenar la información

```python
lista_final = caracteristicas1 + caracteristicas2 + caracteristicas
lista_final
```

4. Análisis utilizando las funciones propias de PLN.

**Tokenización**
Convetir de str a list (para procesar el texto más facilmente)

```python
token=[]
nltk.download('punkt')
from nltk.tokenize import word_tokenize
for i in range(0,15):
  text = lista_final[i].lower() #Pasamos todo a minúsculas
  token = token + word_tokenize(text) #lista donde cada elemento es una palabra o caracter
token
```

```python
#Mirar las palabras que más se repiten y su cantitad de veces
from nltk.probability import FreqDist 
fdist = FreqDist(token)
fdist
```

```python
#Generar una lista con las palabras más repetidas
fdist1=fdist.most_common(15)
fdist1
```

**Lematización:** Evitar duplicación de datos al vincular palabras a la palabra raíz.

```python
from nltk.stem import PorterStemmer 
pst = PorterStemmer() #instanciamos el método
stm = token
token_process=[] #Generamos una lista vacía para que guarde las palabras después de hacer derivación

for word in stm:
  print(word+":"+pst.stem(word))
  token_process.append(pst.stem(word))
```

**Stopwords**

```python
#eliminar palabras vacías
from nltk.corpus import stopwords
nltk.download('stopwords')

palabras_vacias=set(stopwords.words('spanish')) #set:conjunto para que no se repita
#es una base que tiene la librería pero se le pueden añadir más palabras de acuerdo a nuestro objetivo.
palabras_vacias.update(['.',',','(',')',':','|'])
# conservamos las palabras siempre y cuando esa palabra no aparezca en el conjunto de palabras_vacias
text_without_stopwords=[x for x in token if x not in palabras_vacias]

#Realizar nuevamente el proceso de las palabras más repetidas
fdist=FreqDist(text_without_stopwords)
fdist

#Generar una lista con las palabras más repetidas
fdist1=fdist.most_common(20)
fdist1
```

5. **Nube de palabras** sobre el texto.

```python
import matplotlib.pyplot as plt #grafico
from wordcloud import WordCloud, STOPWORDS 
text = ' '.join(lista_final)
wordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords=palabras_vacias, min_font_size = 10).generate(text)
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```
